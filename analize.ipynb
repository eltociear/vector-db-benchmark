```python
import pandas as pd
import json
import plotly.express as px

latencies = []
with open('results/qdrant-default-laion-small-clip-search-0-2023-07-10-10-24-23.json', 'r') as f:
  data = json.load(f)
  latencies = data['results']['latencies']

with open('datasets/laion-small-clip/laion-small-clip/tests.jsonl', 'r') as f:
  conditions = []
  for line in f:
    conditions.append(json.loads(line)['conditions'])
    


df = pd.DataFrame()
df['latency'] = latencies
df['conditions'] = conditions
df.conditions = df['conditions'].apply(lambda x: str(x))
```

```python

fig = px.histogram(df, x="latency", hover_data=["conditions"])
fig.show()
```

```python
fig = px.scatter(df, y="latency", hover_data=["conditions"])
fig.show()
```
